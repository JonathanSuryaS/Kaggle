# Exploratory Data Analysis

Template for Categorical Variable

```python
def plot_categorical(col):
    fig, ax = plt.subplots(1, 2, figsize=(18, 6))
    data[col].value_counts().plot(kind = 'pie', autopct = '%1.1f%%', ax = ax[0])
    data[col].value_counts().plot(kind = 'bar')
    ax[0].set_title(f'Piechart of {col}')
    ax[0].set_ylabel('')
    ax[1].set_title(f'Barchart of {col}')
    plt.show()
```

```python
def plot_numerical(col):
    fig, ax = plt.subplots(1,2, figsize = (18, 6))
    sns.histplot(data = data, x = col, ax = ax[0])
    sns.boxplot(data = data, x = col)
    ax[0].set_title(f'Histogram of {col}')
    ax[1].set_title(f'Boxplot of {col}')
    plt.show()
```

## Categorical Variable

```python
categorical = data.select_dtypes(include=['object']).columns
for col in categorical:
    plot_categorical(col)
```

## Numerical Variable

```python
numerical = data.select_dtypes(include=['number'])
for col in numerical:
    plot_numerical(col)
```

```python
plt.figure(figsize = (30, 15))
sns.heatmap(data = numerical.corr(), annot = True, fmt = '.2f', cmap='YlGnBu')
plt.tight_layout()
plt.show()
```

```python
corr_matrix = numerical.corr().abs()  # Get the absolute values of the correlations
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Identify features with correlation greater than 0.7
to_drop = [column for column in upper.columns if any(upper[column] > 0.7)]
to_drop.pop()
to_drop
```

```python
data = data.drop(columns=to_drop)
print(f"Columns to drop: {to_drop}")
```

# Data Pre Processing

## Handling Missing Values

```python
null_counts = data.isnull().sum()
null_columns = null_counts[null_counts > 0]
print(null_columns)
```

### Dropping Features that have more than 50% missing data

```python
threshold = len(data) * 0.5
data = data.dropna(axis=1, thresh=threshold)
data
```

```python
null_counts = data.isnull().sum()
null_columns = null_counts[null_counts > 0]
print(null_columns)
```

```python
data.drop(columns=['FireplaceQu', 'Id'], inplace = True)
```

### Handling Missing Values

### Lot Frontage

```python
data.groupby(['LotShape', 'LotConfig', 'Street'])['LotFrontage'].mean()
```

```python
data['LotFrontage'] = data['LotFrontage'].fillna(data.groupby(['LotShape', 'LotConfig', 'Street'])['LotFrontage'].transform('mean'))
data['LotFrontage'] = data['LotFrontage'].fillna(data['LotFrontage'].mean())
```

### MasVnrArea

```python
data = data[~data['MasVnrArea'].isna()]
data
```

### BstmtQual

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data
```

### Splitting Feature and Target

```python
X = data.iloc[: , :-1]
y = data.iloc[:, -1]
```

### Encoding Categorical Variables

```python
X = pd.get_dummies(X, drop_first=True)
X = X.astype(float)
X
```

### Splitting Training and Test Set

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, shuffle=True, test_size=0.2)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Modeling

```python
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
mlr = RandomForestRegressor()
mlr.fit(X_train, y_train)
y_pred = mlr.predict(X_test)
mse = r2_score(y_test, y_pred)
mse
```
